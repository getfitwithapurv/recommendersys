```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tdif = TfidfVectorizer(stop_words = 'english') # using TfidVectorizer to transform sentences into arrays of numbers or vectors
df1['Plot']= df1['Plot'].fillna('')
tdif_matrix= tdif.fit_transform(df1['Plot']) #getting a matrix of plots
tdif_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel 
cosine_sim = linear_kernel(tdif_matrix,tdif_matrix) # converting words of plot into vector of matrices to find similarity between vectors
indices = pd.Series(df1.index, index = df1['Title']).drop_duplicates()
```

```python
#here we are creating indices with respect to all movie titles to get recommendations
def get_recommendation(Title,cosine_sim = cosine_sim):
        idx = indices[Title]
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key = lambda X: X[1], reverse = True )
        sim_scores = sim_scores[1:6] #getting recommendation of top 5 movies
        tech_indices = [i[0] for i in sim_scores ]
        return df1['Title'].iloc[tech_indices]
    
```

### Item based Collabarative Filtering 

```python
# Doing this on my earlier dataset
# load the pandas library
import pandas as pd 
# reading the data from my GitHub repository into a Pandas dataframe and saving it in 'reviews'
reviews = pd.read_csv('https://raw.githubusercontent.com/getfitwithapurv/recommendersys/main/review.csv')
reviews.head()
```

```python
matrix = reviews.set_index('Title') #setting title as index
matrix
```

```python
matrix = matrix.fillna(0) #filling the unknown values with 0 
matrix
```

```python
matrix_norm = matrix.subtract(matrix.mean(axis=1), axis = 0) #normalizing the dataset by subtracting the average rating of each movie
matrix_norm.head()
```

```python
# Item similarity matrix using Pearson correlation
item_similarity = matrix_norm.T.corr()
item_similarity.head()
```

```python
# Pick a user 
picked_user = 'Shubham'
# Pick a movie
picked_movie = 'The Batman (2022)'
# Movies that the target user has watched
picked_user_watched = pd.DataFrame(matrix_norm[picked_user].dropna(axis=0, how='all')\
                          .sort_values(ascending=False))\
                          .reset_index()\
                          .rename(columns={'Shubham':'rating'})
picked_user_watched.head()
```

```python
# Similarity score of the picked movie The Batman with all the other movies
picked_movie_similarity_score = item_similarity[[picked_movie]].reset_index().rename(columns={'The Batman (2022)':'similarity_score'})
# Rank the similarities between the movies user Shubham has rated and Gully Boy.
n = 4
picked_user_watched_similarity = pd.merge(left=picked_user_watched, 
                                            right=picked_movie_similarity_score, 
                                            on='Title', 
                                            how='inner')\
                                     .sort_values('similarity_score', ascending=False)[:6]
# Take a look at the User Shubham watched movies with highest similarity
picked_user_watched_similarity
```

```python
# Calculate the predicted rating using weighted average of similarity scores and the ratings from user Shubham
predicted_rating = round(np.average(picked_user_watched_similarity['rating'], weights=picked_user_watched_similarity['similarity_score']), 1)
print(f'The predicted rating for {picked_movie} by user {picked_user} is {predicted_rating}' )
```
