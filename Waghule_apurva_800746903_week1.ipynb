```python
# load the pandas library
import pandas as pd 

# reading the data from my GitHub repository into a Pandas dataframe and saving it in 'reviews'
reviews = pd.read_csv('https://raw.githubusercontent.com/getfitwithapurv/recommendersys/main/reviews%20csv.csv')

reviews.shape  # check the number of rows and columns
```

```python
df = pd.DataFrame(reviews) #adding the data to dataframe for further use
df
```

```python
df = df.drop(df.index[5]) # last row is blank so dropping it 
```

```python
df = df.set_index('Name') # setting the name of reviewers as index
```

```python
df #sanity check
```

```python
mean_movie = (df.sum(numeric_only=True).sum())/sum(df.count()) #finding the mean of all the movie ratings 
mean_movie
```

```python
user_avg = df.mean(axis=1) #adding a new column for avg/mean of ratings for each reviewer
df['User avg'] = user_avg
df
```

```python
df.loc['Movie Avg'] = df.mean() #adding a new row of avg/mean of movies individually 
df
```

```python
df['user avg - mean movie'] = user_avg - mean_movie #adding a new column to get difference of user avg and mean movie
df
```

```python
#adding new row to the dataframe with difference between movie avg and mean movie
new_row = pd.Series(data= df.iloc[5]- mean_movie , name='movie avg - mean movie') 
df = df.append(new_row, ignore_index=False) #append row to the dataframe
df
```

Here we are recommending 'Gully Boy' movie to our reviewer Saurabh using the formula :
Global Baseline Estimate = Mean Movie Rating + Gully boy's rating relative to average + Saurabh's rating relative to average

```python
mean_movie #Mean Movie Rating
```

```python
rra1 = df.iat[6, 4] #Gully boy's rating relative to average
rra1
```

```python
rra2 = df.iat[3, 7] #Saurabh's rating relative to average
rra2
```

```python
G_B_E = mean_movie + rra1 + rra2
G_B_E
```
